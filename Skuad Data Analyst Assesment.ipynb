```python
from selenium import webdriver
import time
import pandas as pd
import os
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options



chrome_options = Options()
chrome_options.add_argument("--headless")  
chrome_options.add_argument("--disable-gpu") 


driver = webdriver.Chrome(options=chrome_options)
options = webdriver.ChromeOptions()
options.add_argument("window-size=1200x600")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url='https://www.linkedin.com/jobs/jobs-in-worldwide?position=1&pageNum=0' 
driver.get(url)



job_type_filter_button = driver.find_element(By.XPATH, '//button[@aria-label="Job type filter. Clicking this button displays all Job type filter options."]')
job_type_filter_button.click()


full_time_checkbox = driver.find_element(By.ID, 'f_JT-0')
contract_checkbox = driver.find_element(By.ID, 'f_JT-2')
full_time_checkbox.click() # for selecting full time job
contract_checkbox.click() # for selecting contract based job

done_button = driver.find_element(By.XPATH, '/html/body/div[1]/section/div/div/div/form/ul/li[3]/div/div/div/button')
done_button.click()


remote_filter_button = driver.find_element(By.XPATH, '//button[@aria-label="On-site/remote filter. Clicking this button displays all On-site/remote filter options."]')
remote_filter_button.click()

remote_option = driver.find_element(By.ID, 'f_WT-2')
remote_option.click() # for selecting remote jobs only

done_button = driver.find_element(By.XPATH, '/html/body/div[1]/section/div/div/div/form/ul/li[5]/div/div/div/button')
done_button.click()

# As it is is mentioned in assignment that i have to extract the job details only for full time and contract based and should be remote


wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

i = 2
while i <= 2000:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i = i + 1  # for continue scrolling
    
    # this try except code handle the "see more job" button 
    # in one time it only shows 25 job for more jobs we have to click on this button
    try:
        send=driver.find_element(By.XPATH, "//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();", send)   
        time.sleep(3)
    except:
        pass
        time.sleep(5)


# this is for Company Name
company_elements = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')
company_names = []
for company_element in company_elements:
    company_names.append(company_element.text)


# this is for Job Title
title_elements = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
job_titles = []
for title_element in title_elements:
    job_titles.append(title_element.text)
    

# this is for Job Location
job_elements = driver.find_elements(By.CLASS_NAME, 'job-search-card__location')
job_locations = []
for job_element in job_elements:
    job_locations.append(job_element.text)
    
    
    
# this is for job benefits
b_elements = driver.find_elements(By.CLASS_NAME, 'result-benefits')
benefits=[]
for level in b_elements:
    benefits.append(level.text)
    
    
'''a_elements = driver.find_elements(By.CLASS_NAME, 'job-search-card__listdate--new')
job_date=[]
for level in a_elements:
    job_date.append(level.get_element('datetime').text)
'''
   
 # this is for Job Post Date   
a_elements = driver.find_elements(By.CLASS_NAME, 'job-search-card__listdate--new')
job_date = []
for element in a_elements:
    job_date.append(element.get_attribute('datetime'))
    
  
 # this is for Job link
jobList = driver.find_elements(By.CLASS_NAME, 'base-card__full-link')
hrefList = []
for e in jobList:
    hrefList.append(e.get_attribute('href'))

driver.quit()
```
